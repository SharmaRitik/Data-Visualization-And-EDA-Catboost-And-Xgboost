```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train = pd.read_csv('train_values.csv')
target = pd.read_csv('train_labels.csv')
test = pd.read_csv('test_values.csv')
```

Lets combine train and target together

```python
df_train = train.merge(target, how = 'left', on = 'building_id')
```

```python
df_train.nunique(dropna=False)
```

Now Lets find NaN's values in df_train

```python
def null_analysis(df):
  '''
  desc: get nulls for each column in counts & percentages
  arg: dataframe
  return: dataframe
  '''
  null_cnt = df.isnull().sum() # calculate null counts
  null_cnt = null_cnt[null_cnt!=0] # remove non-null cols
  null_percent = null_cnt / len(df) * 100 # calculate null percentages
  null_table = pd.concat([pd.DataFrame(null_cnt), pd.DataFrame(null_percent)], axis=1)
  null_table.columns = ['counts', 'percentage']
  null_table.sort_values('counts', ascending=False, inplace=True)
  return null_table

null_analysis(df_train)
```

No NaN's values in the dataset

```python
df_train = df_train.drop(['building_id'],axis=1)
test = test.drop(['building_id'],axis=1)

## Lets extract the target and also drop the target variable from the training dataset
y = df_train.loc[:,'damage_grade']
df_train = df_train.drop('damage_grade',axis=1)
```

```python
## Lets first find out the statistical information about feature
df_train.describe()
```

```python
## Lets first look at if there is a class imbalance
ax = sns.countplot(y,label='Count')
A,B,C = y.value_counts()
print("No of class 1",A)
print("No of class 2",B)
print("No of class 3",C)
```

```python
#correlation map
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(df_train.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax);
```

```python
## Lets drop columns having 80% correlation
drop_cols = ['height_percentage','count_floors_pre_eq']
df_train = df_train.drop(drop_cols,axis=1)
test = test.drop(drop_cols,axis=1)
```

Lets encode categorical variable

```python
## Lets first encode categorical variable
from category_encoders.cat_boost import CatBoostEncoder
CBE_encoder = CatBoostEncoder()
cat_cols = list(df_train.select_dtypes(include=['object']).columns)
train_cbe = df_train.copy()
test_cbe = test.copy()
train_cbe[cat_cols] = CBE_encoder.fit_transform(train[cat_cols], y)
test_cbe[cat_cols] = CBE_encoder.transform(test[cat_cols])
```

Lets split the dataset into train,validation

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_cbe, y, test_size=0.20, random_state=0,shuffle = True)
```

As we have seen that there is a class imbalance problem in the dataset so lets fix it

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()

x_train_smote , y_train_smote = smote.fit_sample(X_train,y_train)
```

```python
# find out the indices of categorical variables
categorical_var_train = np.where(x_train_smote.dtypes == object )[0]
categorical_var_test = np.where(X_test.dtypes == object)[0]
```

```python
from catboost import Pool
train_pool = Pool(data=x_train_smote, label=y_train_smote,cat_features = categorical_var_train)
test_pool = Pool(data=X_test, label=y_test.values , cat_features = categorical_var_test) 
```

# Now we will fit our catboost classifier to our dataset

```python
from catboost import  CatBoostClassifier

model = CatBoostClassifier(
    iterations=3500,
    random_strength=0.1,
    loss_function= 'MultiClass',
    eval_metric='TotalF1',
    task_type='CPU',
    learning_rate = 0.3,
    depth = 8,
    l2_leaf_reg = 9
    
)
```

```python
model.fit(train_pool,eval_set=test_pool)
```

```python
pred_cat = model.predict(X_test)
from sklearn.metrics import f1_score
score_catboost = f1_score(pred_cat,y_test,average='micro')
```

## Lets  try LightGbM 

```python
cat_cols = list(df_train.select_dtypes(include=['object']).columns)
```

```python
from category_encoders.cat_boost import CatBoostEncoder
CBE_encoder = CatBoostEncoder()
train_cbe = df_train.copy()
test_cbe = test.copy()
train_cbe[cat_cols] = CBE_encoder.fit_transform(df_train[cat_cols], y)
test_cbe[cat_cols] = CBE_encoder.transform(test[cat_cols])
```

```python

X_train_gbm, X_test_gbm, y_train_gbm, y_test_gbm = train_test_split(train_cbe, y, test_size=0.20, random_state=0,shuffle = True)

x_trainlgbm_smote , y_trainlgbm_smote = smote.fit_sample(X_train_gbm,y_train_gbm)
```

```python
#Converting the dataset in proper LGB format
import lightgbm as lgb
d_train=lgb.Dataset(x_trainlgbm_smote, label=y_trainlgbm_smote)
#setting up the parameters
params={}
params['learning_rate']=0.03
params['boosting_type']='gbdt' #GradientBoostingDecisionTree
params['objective']='multiclassova' #Multi-class target feature
params['metric']='multi_logloss' #metric for multi-class
params['max_depth']=10
params['num_class']=3 #no.of unique values in the target class not inclusive of the end value
params['bagging_freq'] = 9
params['feature_fraction'] = 0.4
```

```python
clf=lgb.train(params,d_train,500)
```

```python
pred_LGBM = clf.predict(X_test_gbm)
```

```python
y_pred = [np.argmax(line) for line in pred_LGBM]
```

```python
from sklearn.metrics import f1_score
score_gbm = f1_score(y_pred,y_test_gbm,average = 'micro')
```
